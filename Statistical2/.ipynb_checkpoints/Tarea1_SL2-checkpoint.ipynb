```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling

from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.datasets import fetch_openml

from sklearn.decomposition import PCA
from sklearn import datasets
import imageio as imio
from sklearn.manifold import TSNE
import time
```

### Data, Images, Labels, Features

```python
images_inception_features = './pre-processing/training_inception_features_array.npy'
training_image_labels_array = './pre-processing/training_image_labels_array.npy'
training_image_names_array = './pre-processing/training_image_names_array.npy'
```

```python
TRAIN_DICTIONARY_KEYS = ["images", "label","features"]

training_set_dict = dict()
    
training_set_dict['images_name'] = np.load(training_image_names_array,allow_pickle=True)
training_set_dict['label'] = np.load(training_image_labels_array,allow_pickle=True)
training_set_dict['features'] = np.load(images_inception_features,allow_pickle=True)
```

```python
features = training_set_dict['features']
labels = training_set_dict['label']
images_name = training_set_dict['images_name']
```

```python
print('Label: ',labels[4])

print('Nombre: ',images_name[4])

img = imio.imread('./data/train/'+images_name[4]+'.jpg')
plt.imshow(img)
```

### PCA

```python
time_start = time.time()
pca = PCA(n_components = 2)
X2D = pca.fit_transform(features)
print ('PCA listo. Tiempo transcurrido:' +format(time.time()-time_start)+ 'segundos')
```

### T SNE

```python
time_start = time.time()
X2D_TSNE = TSNE(n_components=2).fit_transform(features)
print ('TSNE listo. Tiempo transcurrido:' +format(time.time()-time_start)+ 'segundos')
```

```python
print(features.shape)
print(X2D.shape)
```

```python
X2D_TSNE.shape
```

### Gráficas PCA vs T SNE

```python
plt.scatter(X2D[:, 0], X2D[:, 1])
plt.xlabel('component 1')
plt.ylabel('component 2')
```

```python
plt.scatter(X2D_TSNE[:, 0], X2D_TSNE[:, 1])
plt.xlabel('component 1')
plt.ylabel('component 2')
```

### KMeans Cluster

```python
kmeans = KMeans(n_clusters=10, random_state=0).fit(X2D)
```

```python
clusters = kmeans.predict(X2D)
clusters
```

```python
kmeans.cluster_centers_.shape
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
numlabels=np.arange(len(labels))
```

```python
plt.scatter(X2D[:, 0], X2D[:, 1],
            c=numlabels, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('Accent', 10))
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

### TSNE VS PCA
* En primer lugar esta el tiempo de ejecucuión que es por bastante diferencia PCA más rápido
* TSNE puede trabajar con modelos no lienales, mientras que PCA solo con modelos lineales, por lo tanto se pierde mucha información
* TSNE tiene baja interpretación de su algoritmo, por lo que permite visualizar muy bien los datos, pero no genera, ecuaciones fácilmente interpretables como lo hace PCA.
