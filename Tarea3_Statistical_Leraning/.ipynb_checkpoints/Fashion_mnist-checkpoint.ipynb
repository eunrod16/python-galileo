```python
label_dict = {
        0: 'T - shirt / top',
        1: 'Trouser',
        2: 'Sandal'
}
```

```python
#prepare data test
selection_test = (test_labels[:] == 0) | (test_labels[:] == 1) | (test_labels[:] == 5)
selection_test
test_labels_select = test_labels[selection_test]
test_images_select = test_images[selection_test]
```

```python
#prepare data train
selection_train = (train_labels[:] == 0) | (train_labels[:] == 1) | (train_labels[:] == 5)
selection_train
train_labels_select = train_labels[selection_train]
train_labels_select.shape
train_labels_select[0:10]
train_images_select = train_images[selection_train]
train_images_select.shape
train_reshape =  [np.reshape(x,-1) for x in train_images_select]
train_x = np.divide (train_reshape,255)
#train_x = train_reshape
```

```python
#onehotcoding
a = np.unique(train_labels_select, return_inverse=1)[1]
target = (a.ravel()[:,None] == np.arange(a.max()+1)).astype(int)
target
target.shape
```

```python
m = 784
n = 3
epochs = 100
tf.reset_default_graph()
with tf.name_scope('datos_entrenamiento'):
    x = tf.placeholder(tf.float32, [None, m], name = "X") 
    y = tf.placeholder(tf.float32, [None, n], name = "Y") 
with tf.name_scope('parametros'):
    W = tf.Variable(tf.zeros([m,n]), name = "W") 
    b = tf.Variable(tf.zeros([n]), name = "B")
with tf.name_scope('modelo'):
    y_ = tf.nn.softmax(tf.matmul(x, W) + b)
with tf.name_scope('cross_entropy'):
    cross_entropy = tf.reduce_mean(-tf.reduce_sum(y*tf.log(y_), reduction_indices=[1]))
    train_step = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy) 
    tf.summary.scalar('cost',cross_entropy)
with tf.name_scope('accuracy'):
    correct_prediction = tf.equal(tf.argmax(y_,1), tf.argmax(y,1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    tf.summary.scalar('accuracy',accuracy)

summaries = tf.summary.merge_all()
```

```python
saver = tf.train.Saver()
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    summary_writer = tf.summary.FileWriter('./graphs/',sess.graph)
    batch_size = 32
    batchs = int(np.around(len(train_x)/batch_size))
    batchs
    for i in range(5):
        i0 = 0
        i1 = batch_size-1
        for batch in range(batchs):
            x_batch = train_x[i0:i1]
            y_batch = target[i0:i1]
            accurracy_,cross_entropy_, _,summary = sess.run([accuracy,train_step, y_,summaries], feed_dict={x: x_batch,
                                           y: y_batch})
            if batch%100==0:
                summary_writer.add_summary(summary,i)
                print ('Epoch: ',i,'Batch:' ,batch, ' accuracy: ', accurracy_,' output example:',_[0])
            i0 = i0+batch_size
            i1 = i1+batch_size

    save_path = saver.save(sess, "session.ckpt")
summary_writer.close()
```

```python
def WhatClotheIs(arreglo):
    test_reshape =  np.reshape(arreglo, (1, 784))
    test_x = np.divide (test_reshape,255)
    with tf.Session() as sess:
        saver.restore(sess, "session.ckpt")
        predict = sess.run([y_], feed_dict = {x: test_x})
        categoria = np.argmax(predict)
        plt.figure()
        plt.imshow(arreglo, cmap = 'gray')
        plt.xlabel(label_dict[categoria] + ' con probabilidad ' + str(round(np.max(predict) * 100,2)) + '%')
        plt.colorbar()
WhatClotheIs(test_images_select[22])

test_labels_select[22]
```
