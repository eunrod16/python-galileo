```python
#onehotcoding
a = np.unique(train_labels_s, return_inverse=1)[1]
target = (a.ravel()[:,None] == np.arange(a.max()+1)).astype(int)
target
target.shape
```

```python
train_images_s = train_images[selection_train]
train_images_s.shape
train_reshape =  [np.reshape(x,-1) for x in train_images_s]
train_reshape[0]
```

```python
train_x = np.divide (train_reshape,255)
np.shape(train_x)
train_x[0]
```

```python
m = 784
n = 3
epochs = 100
x = tf.placeholder(tf.float32, [None, m]) 
#  n = al numero de posibles categorias
y = tf.placeholder(tf.float32, [None, n]) 

# tip: W es una matriz
W = tf.Variable(tf.zeros([m,n])) 
b = tf.Variable(tf.zeros([n]))

y_ = tf.nn.softmax(tf.matmul(x, W) + b)

cross_entropy = tf.reduce_mean(-tf.reduce_sum(y*tf.log(y_), reduction_indices=[1]))
train_step = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy) 
tf.summary.scalar('cost',cross_entropy)

correct_prediction = tf.equal(tf.argmax(y_,1), tf.argmax(y,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
tf.summary.scalar('accuracy',accuracy)

summaries = tf.summary.merge_all()


# la ejecucion(sesion) hacerla con mini-batch gradient descent con un batch size = 32
# cada cierto numero de iteraciones "n" imprimir: el numero de iteracion, el loss de la iteracion, el accuracy de la iteracion y "enviarlo" a tensorboard
```
